```python
import cv2
import matplotlib.pyplot as plt

from random import randint
```

```python
img = cv2.imread('imgs/121.jpg', 1)
img = img[:,:,::-1]

plt.rcParams['figure.figsize'] = [10, 10]
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
for _ in range(30):
    random_x = randint(0, 462) # We will use a 10x10 area
    random_y = randint(0, 462)
    
    img[random_x:random_x+100, random_y:random_y+100] = 255

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
g_blur = cv2.GaussianBlur(img,(111,111),0)
m_blur = cv2.medianBlur(img, 151)
blur = cv2.blur(img, (100,100))

images = [g_blur, m_blur, blur]
titles = ['gaussian', 'median', 'blur']
plt.rcParams['figure.figsize'] = [20, 20]

for i in range(3):
    plt.subplot(1,3,i+1), plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
